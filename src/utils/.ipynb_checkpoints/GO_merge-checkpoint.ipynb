```python
import json
import itertools
import ast
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import sys, os
import traceback
from itertools import chain
import numpy as np
import pandas as pd
```

```python
def process_data(rootdir, out_file):
    of = open(out_file,'w')
    for subdir, dirs, files in os.walk(rootdir):
        for fname in files:
            path1 = subdir + fname
            path2 = '/'.join(rootdir.split('/')[:-2]) + '/output_go_bpmf/' + fname
            print path1
            print path2
            if 'DS_Store' not in path1 and 'DS_Store' not in path2:
                with open(path1) as f1, open(path2) as f2:
                    for f1_line1, f2_line1 in itertools.izip(f1, f2):
                        try:
                            f1_line2 = next(f1)
                            f2_line2 = next(f2)
                            txt1 = ast.literal_eval(f1_line1)
                            txt2 = ast.literal_eval(f2_line1)
                            lab1 = ast.literal_eval(f1_line2)
                            lab2 = ast.literal_eval(f2_line2)
                            lab = []
                            for i in range(len(lab1)):
                                if lab1[i] == 'O' and lab2[i] == 'O':
                                    lab.append('O')
                                elif lab1[i] != 'O':
                                    lab.append(lab1[i])
                                else:
                                    lab.append(lab2[i])
                            of.write(str(txt1) + '\n')
                            of.write(str(lab) + '\n')
                        except:
                            pass
        
    of.close()
    
```

```python
process_data('../../data/input/GO/output_go_cc/', '../../data/input/GO/merged/merged_file.txt')
```
